```python
### TEST FUNCTION: test_question1
import numpy as np
import scipy.io

# Task 1: Feature Extraction
def extract_features(data):
    """
    Extracts two features from the dataset:
    - Feature 1: Average brightness
    - Feature 2: Standard deviation of brightness
    """
    feature1 = np.mean(data, axis=(1, 2))  # Average brightness
    feature2 = np.std(data, axis=(1, 2))   # Standard deviation
    return np.column_stack((feature1, feature2))  # Combine features into 2D array

# Task 2: Parameter Calculation
def calculate_parameters(features):
    """
    Calculates the mean and variance for both features.
    Returns:
    - mean_f1: Mean of Feature 1
    - var_f1: Variance of Feature 1
    - mean_f2: Mean of Feature 2
    - var_f2: Variance of Feature 2
    """
    mean_f1 = np.mean(features[:, 0])
    var_f1 = np.var(features[:, 0])
    mean_f2 = np.mean(features[:, 1])
    var_f2 = np.var(features[:, 1])
    return mean_f1, var_f1, mean_f2, var_f2

# Task 3: Naive Bayes Classifier
def gaussian(x, mean, variance):
    """
    Gaussian probability density function.
    """
    return (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-((x - mean) ** 2) / (2 * variance))

def naive_bayes(features, mean0, var0, mean1, var1):
    """
    Classifies data points using Naive Bayes.
    - features: 2D features array
    - mean0, var0: Parameters for class 0
    - mean1, var1: Parameters for class 1
    Returns:
    - Predicted labels (0 or 1)
    """
    p0 = gaussian(features[:, 0], mean0[0], var0[0]) * gaussian(features[:, 1], mean0[1], var0[1])
    p1 = gaussian(features[:, 0], mean1[0], var1[0]) * gaussian(features[:, 1], mean1[1], var1[1])
    return (p0 <= p1).astype(int)  # Predict 1 if p1 >= p0, else 0

# Task 4: Accuracy Calculation
def calculate_accuracy(predictions, true_labels):
    """
    Calculates accuracy by comparing predictions with true labels.
    """
    return np.mean(predictions == true_labels)

# Main function
def main():
    myID = '0900'  # Replace with your ASU ID's last 4 digits

    # Load data
    Numpyfile0 = scipy.io.loadmat('digit0_stu_train' + myID + '.mat')
    Numpyfile1 = scipy.io.loadmat('digit1_stu_train' + myID + '.mat')
    Numpyfile2 = scipy.io.loadmat('digit0_testset' + '.mat')
    Numpyfile3 = scipy.io.loadmat('digit1_testset' + '.mat')
    train0 = Numpyfile0.get('target_img')
    train1 = Numpyfile1.get('target_img')
    test0 = Numpyfile2.get('target_img')
    test1 = Numpyfile3.get('target_img')

    # Task 1: Feature extraction
    train0_features = extract_features(train0)
    train1_features = extract_features(train1)
    test0_features = extract_features(test0)
    test1_features = extract_features(test1)

    # Task 2: Parameter calculation
    mean_f1_0, var_f1_0, mean_f2_0, var_f2_0 = calculate_parameters(train0_features)
    mean_f1_1, var_f1_1, mean_f2_1, var_f2_1 = calculate_parameters(train1_features)

    # Task 3: Classification
    test0_pred = naive_bayes(test0_features, [mean_f1_0, mean_f2_0], [var_f1_0, var_f2_0],
                             [mean_f1_1, mean_f2_1], [var_f1_1, var_f2_1])
    test1_pred = naive_bayes(test1_features, [mean_f1_0, mean_f2_0], [var_f1_0, var_f2_0],
                             [mean_f1_1, mean_f2_1], [var_f1_1, var_f2_1])

    # Task 4: Accuracy calculation
    accuracy_test0 = calculate_accuracy(test0_pred, np.zeros(len(test0_features)))
    accuracy_test1 = calculate_accuracy(test1_pred, np.ones(len(test1_features)))

    # Final output
    output = [
        myID,
        float(mean_f1_0), float(var_f1_0), float(mean_f2_0), float(var_f2_0),
        float(mean_f1_1), float(var_f1_1), float(mean_f2_1), float(var_f2_1),
        float(accuracy_test0), float(accuracy_test1)
    ]

    print(output)

if __name__ == '__main__':
    main()
```
